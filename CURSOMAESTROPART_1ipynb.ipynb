```python
# Digamos que quieres hacer una calculadora, con las funciones y operaciones basicas
import math

num1 = int(input("Ingresa el primer numero: ")) # pregunta a traves de un input por consola
num2  = int(input("ingrese el segundo numero: "))
operaciones = input(" Ingrese la operacion que desea realizar 1: suma,  2: resta, 3: multiplicación, 4: División, 5: potenciación, 6: radicación ")

# condicionales
if operaciones == "1":
    suma = num1 + num2
    print(f"El resultado de la suma es: {suma}")
elif operaciones == "2":
  resta = num1 - num2
  print(f"El resultado de la resta es: {resta}")
elif operaciones == "3":
  multi = num1 * num2
  print(f"El resultado de la multiolicacion  es: {multi}")
elif  operaciones == "4":
  division = num1 / num2
  print(f"El resultado de la divi es:{division}")
elif operaciones == "5":
  potencia = num1 ** num2
  print(f"El resultado de la potencia es: {potencia}")
elif operaciones == "6":
  numraiz =  int(input("Ingrese el numero de cual quiere saber la raiz"))
  raiz = math.sqrt(numraiz)
  print(f"El resultado de la raiz es: {raiz}")
else:
  print("Operación no válida")
```

```python
# slicing en las cadenas
# el slicing se entiende, como capacidad de las cadenas que
# devuelven un subconjunto o una cadena utilizando dos indices

palabra1 = "Jxulianea"
print(palabra1[1:-1])
print(palabra1[3:])

# si con slicing no se le indica el indice se toma por defecto el principio y el final
palabra1[:]

palabra1[-1]  + palabra1[3] + palabra1[-2] + palabra1[1]
```

```python
# si un indice esta fuera de la cadena

palabra1[99]
```

```python
# pero el slicing no pasa esto, lo ignora basicamente

palabra1[:100]
palabra1[100:]
```

```python
# inmutabilidad
# una propiedad de las cadenas es que no se pueden modificar, si intentamos reasignar un caracter, no nos dejara
palabra1[0] = "B"
```

```python
# sin embargo mediante la implementación de slicing y concatenación podemos generar
# nuevas cadenas facilmente

palabra1
palabra1 = "u" + palabra1[1:]
palabra1
```

```python
# Funciones
# Un buen ejemplo de una función util para nosotros o
# para el manejo general de caracteres es len() #Saber la longitud, numeros de carcteres

len(palabra1)

CADENATOTAL = "BUENAS TARDES VAMOS A PROBAR ESTA PEQUEÑA CADENA"
len(CADENATOTAL)
```
